### Exercise 1 

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7

We could use a mathematical equation to solve this or we can run the experiment many times, and then count the number of times where the score is even or greater than 7

-Run the experiment 1000 times (roll 2 dice 1000 times, and sum the result)

-Keep track of the number of times that the sum was either greater than 7 or an even number

-Divide the number from step 2 by the number of iterations (1000)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

def mc_pi_aprox(N=10000):    
    x, y = np.random.uniform(0, 1, size=(2, N))
    interior = (x**2 + y**2) <= 1
    exterior = np.invert(interior)


    plt.plot(x[interior], y[interior], 'b.')
    plt.plot(x[exterior], y[exterior], 'r.')

    pi = interior.sum() * 4 / N
    error = abs((pi - np.pi) / pi) * 100

    plt.plot(0, 0, label='N={:d}\n$\hat\pi$ = {:4.4f}\nerror = {:4.4f}%'
             .format(N,pi,error), alpha=0.1)
    plt.axis('square')
    plt.legend(frameon=True, framealpha=0.9, fontsize=10, loc="upper right")
```

```python
plt.figure(figsize=(8,8))  # tamaño de la figura

plt.subplot(2,2,1)
mc_pi_aprox(N=100)
plt.subplot(2,2,2)
mc_pi_aprox(N=500)
plt.subplot(2,2,3)
mc_pi_aprox(N=1000)
plt.subplot(2,2,4)
mc_pi_aprox(N=1000000)

plt.show()
```

```python
np.random.seed(42)

def mi_ecuacion_aprox(N=10000): 

    x_max = 1
    x_min = -1
    x_range = x_max-x_min
    x = np.random.uniform(x_min, x_max, size=(1, N))
    y = np.random.uniform(0, 1, size=(1, N))

    interior = (x**2) >= y
    exterior = np.invert(interior)


    plt.plot(x[interior], y[interior], 'b.')
    plt.plot(x[exterior], y[exterior], 'r.')

    integral = interior.sum()*x_range / N
    error = abs((integral - 1/3.0) / integral) * 100

    plt.plot(0, 0, label='N={:d}\n$\int_0^1 x^2 dx $ = {:4.4f}\nerror = {:4.4f}%'
             .format(N,integral,error), alpha=0.1)
    #plt.axis('square')
    plt.legend(frameon=True, framealpha=0.9, fontsize=10, loc="upper right")
```

```python
plt.figure(figsize=(8,8))  # tamaño de la figura

plt.subplot(2,2,1)
mi_ecuacion_aprox(N=100)
plt.subplot(2,2,2)
mi_ecuacion_aprox(N=500)
plt.subplot(2,2,3)
mi_ecuacion_aprox(N=1000)
plt.subplot(2,2,4)
mi_ecuacion_aprox(N=10000)

plt.show()
```
